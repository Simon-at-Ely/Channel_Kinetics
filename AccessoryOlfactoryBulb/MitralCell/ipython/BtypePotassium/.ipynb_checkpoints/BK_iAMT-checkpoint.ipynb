#BK iAMC Potassium channel model
**Our starting point for this model is the Maex and De Schutter 1998 Granule cell KCa channel:https://senselab.med.yale.edu/ModelDB/ShowModel.cshtml?model=50219. 

```python
%pylab inline
ca_conc = 0.025
v = np.linspace(-0.1, 0.1, 2000)
malpha = 2500/(1 + ( (1.5e-3 *(np.exp(-85*v))) / ca_conc))
mbeta = 1500/(1 + (ca_conc / (1.5e-4 * (np.exp (-77*v)))))
plt.plot(v, malpha, v, mbeta)
```

```python
mtau=1/(malpha + mbeta)
plt.plot(v,mtau)
```

```python
minf = malpha/(malpha + mbeta)
plt.plot(v,minf)
```
