##Compartmental Modelling

**The model design and specification was set up in "neuroConstruct" software (Gleeson et al. 2007). Model scripts were generated at mutiple stages during the development process by "neuroConstruct" to run simulation using the "Neuron" simulator software (Hines and Carnevale 1997, 2000).** 

##Morphological Reconstruction

**Digitised cell reconstructions were produced using...details to be added... A single representative cell morphology of an intrinsically oscilating accessory olfactory bulb mitral cell was chosen for the development of the model (see Figure above). The morphology parameters were saved as a .hoc file that was imported into "neuroConstuct". Some editing was needed to separate compartments into groups (soma, dendrites, axon) for specifying channel allocation.**  

##Fitting the Passive Parameters

**The specific passive parameters of the model were tuned in the model to target mean experimental measurements/estimates. The experimental parameters used for this process were somatic input resistance ($I_{Input}$), membrane time constant (${\tau}_m$), membrane capcitance ($C_m$). For each tuning iteration a set of specific parameters were set in the model for all compartments in the morphology reconstruction. A simulated current injection of 70 picoAmperes  was applied to the soma for 400 ms and $I_{Input}$ was calculated from the difference between starting membrane potential and steady state over the amplitude of the current injection. While ${\tau}_m$ was estimated by fitting an exponential to the hyperpolarisation transient from near starting membrane potential to the steady state (using the "Octave expfit" function http://octave.sourceforge.net/optim/function/expfit.html). Then a simulated voltage clamp was applied to the soma of the model (5 ms 5 mV, 5 ms 5 mV, 20 $M{\Omega}$. $C_m$ was estimated by fitting an exponential to the second transient (again with "Octave expfit"). Iterations were repeated until the specific passive parameters set in the model replicated the experimental parameters.**

##Active Properties

**Four new channel models are included in this project. They have been developed from IV plots, steady state plots and time course data sets for activation, inactivation, and deactivation from voltage clamp experiments. The analysis to produce these data sets have been carried out with IGOR in the first place with further data fitting in ipython notebooks using pandas, numpy and scipy.**

###Fast Transient Sodium Channel (Nax, from Migliore et al. 2005).

$G_{Nax}=G_{max}*m^3*h$

$G_{max}= 7.0E-8 mS {\mu}m^{-2}$ Soma, Dendrite, Axon (7.0E-7 zero shunt version)

**Physiological Units (mV and ms)**

$Q_{10}=2^{(\frac{(t - 24)}{10})}$

**Activation**

$m_{\alpha}(v) = \frac{2.880000018 * \frac{(v + 20)}{7.2}}{1- exp -(\frac{(v + 20)}{7.2})}$

$m_{\beta}(v) = \frac{0.892800005 * \frac{(v + 20)}{-7.2}}{1- exp -(\frac{(v + 20)}{-7.2})}$

${\tau}_m(v) = \frac{1}{((m_{\alpha} + m_{\beta}) * Q_{10})}$ 

 If... ${\tau}_m(v) = 0.02$; ${\tau}_m(v) = (0.02 * Q_{10})$ 
    
 Else... ${\tau}_m(v) = \frac{1}{(m_{\alpha} + m_{\beta})}$

**Inactivation**

$h_{\alpha}(v) = \frac{0.045 * \frac{(v + 35)}{1.5}}{1- exp -(\frac{(v + 35)}{1.5})}$

$h_{\beta}(v) = \frac{0.015 * \frac{(v + 35)}{-1.5}}{1- exp -(\frac{(v + 35)}{-1.5})}$

${\tau}_h(v) = \frac{1}{((h_{\alpha} + h_{\beta}) * Q_{10})}$ 

If... ${\tau}_h(v) = 0.5$; ${\tau}_h(v) = (0.5 * Q_{10})$ 

Else... ${\tau}_h(v) = \frac{1}{(h_{\alpha} + h_{\beta})}$

$h_{\infty}(v) = \frac{1}{1+ exp (\frac{(v + 40)}{4})}$
