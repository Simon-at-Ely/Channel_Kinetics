**This was converted to Steady State ($m_{\infty}$) using the relationship** $I=G^*(v-ENa)$ where $G = Gmax^*m_{\infty}$

Thank-you to Volker Steuber for suggesting this approach..

````python
#load python packages
%pylab inline

#set parameters
vhalf = -60.563
scale = 5.2909
base = 0.41431
vmax = 108.69
i_exp = array([14.03, 33.44, 58.98, 79.38])
v_exp = array([-70, -65, -60,-55])
v = np.linspace(-90, 40, num=2000)

#convert to steady state
ENa = 67
Gmax = 109/(-40-ENa)
df_exp = v_exp-ENa
G_exp = i_exp/df_exp
minf_exp = G_exp/Gmax
G_fit = (base + (vmax/(1+exp((vhalf-v)/scale))))/(v-ENa)
minf_fit = G_fit/Gmax

#plot I/V curve
plt.plot(v, minf_fit, v_exp, minf_exp, 'o')
plt.title('iAMC_NaP steady state activation')
plt.xlabel('Vmemb (mV)')
plt.ylabel('minf(dimensionless)')
plt.legend(['fit', 'experimental'], loc='lower right')
````
